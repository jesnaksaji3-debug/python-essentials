**This file contains examples of break, continue and matrix handling**

```python
#break statement
for i in range(6):
  if i == 4:
    #break #here the loop breaks when i becomes 4 and doesn't continue
    continue #here the loop skips 4
  print(i)
```

**List of list**

```python
#to handle the marks of 4 subjects for 3 students
marks = [
    [50,30,40,45],
    [80,40,55,67],
    [55,45,65,75]
]
print(marks)
print(marks[0])
print(marks[2][2])
for i in range(3):
  print(f"Marks of student {i+1}")
  for j in range(4):
    print(marks[i][j])
```

**Identity operator**

```python
#is and is not
marks_stud1 = [23,45,67]
marks_stud2 =  [23,45,67]
x = marks_stud2
print(marks_stud1 == marks_stud2)
print(marks_stud1 is marks_stud2)
print(x == marks_stud1)
print(x is marks_stud2) #memory allocated is different
print(x is marks_stud1)
```
